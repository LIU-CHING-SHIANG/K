```python
import yfinance as yf
import pandas as pd

# 定义股票代码和时间范围
tickers = ['GOOG', 'AMZN', 'MSFT', 'AAPL', 'META']  # Facebook 现在的代码是 META
start_date = '2014-01-01'
end_date = '2024-11-14'

# 创建一个空的数据框用于存储合并后的数据
data_frames = []

# 获取每个股票的数据
for ticker in tickers:
    df = yf.download(ticker, start=start_date, end=end_date)
    # 添加股票代码作为列的顶层标签
    df.columns = pd.MultiIndex.from_product([[ticker], df.columns])
    data_frames.append(df)

# 合并所有股票的数据
combined_df = pd.concat(data_frames, axis=1)

# 显示结果
print(combined_df)
```

```python
combined_df.columns.names = ['股票代號', '股票資訊']
```
